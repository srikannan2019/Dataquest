```python
import pandas as pd
import numpy as np
import matplotlib as plt
test_logs = pd.read_csv(r"C:\Users\SRIVIDHYA\Desktop\ArkoseLabs\arkose_labs_DS_test_logs.csv")
test_logs.head()

```

The file is  extracted as csv to a dataframe and initial view to file shows there are 'Nan" values for the columns "Unnamed 4 to 29".So drop those columns for better analysis of the data set.So the new data frame has only 4 columns that is needed for our analysis.


```python
new_test_log = test_logs.drop(test_logs.loc[:,'Unnamed: 4':'Unnamed: 29'].head(0).columns, axis=1)
```

```python

new_test_log.shape
```

The data frame has 4 columns and 1779 rows .The column 'solved' has null values to be handled.The message and solved are related like if the message is "user submitted answer' is '1' for solved  for other messages it is 'Nan'.So only if the user submitted correct answer is needed to  find the completion rate.so we can fill the other message columns as '0' since only the right answers  with user submitted are going to be considered for the completion rate.

```python

new_test_log.isnull().sum()
```

```python
new_test_log.head(30)
```

```python
new_test_log.loc[new_test_log["message"] == "game loaded", "solved"] = 0.0
new_test_log.loc[new_test_log["message"] == "user clicked verify", "solved"] = 0.0
new_test_log.loc[new_test_log["message"] == "user refreshed", "solved"] = 0.0
```

So now the data set is clean and ready to find the completion rate of A and B with respect to correct answers.

```python
print(new_test_log['solved'].value_counts(dropna = False))
```

```python
new_test_log.head(30)
```

Select only sorted  column with 1 as values to find the completion rate for A and B


```python
new_test_log['theme_ab'].value_counts()
```

```python
if [(new_test_log['solved'] == 1) &  (new_test_log['theme_ab'] == 1)]:
    completion_rate_A = new_test_log['theme_ab'].value_counts(normalize= True).sort_index() * 100
elif [(new_test_log['solved'] == 1) &  (new_test_log['theme_ab'] == 2)]: 
    completion_rate_B = new_test_log['theme_ab'].value_counts(normalize= True).sort_index() * 100
completion_rate_A
completion_rate_B 
```

The completion rate for users completing the challenge with respect to A is 61.7% and B is 38.27 %.The result of this analysis is in the result_test_log file
