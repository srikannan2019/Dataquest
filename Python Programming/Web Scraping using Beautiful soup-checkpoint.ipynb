```python
#Import beautiful soup and request libraray in conda prompt
#now we are going to web scraping for a wikipedia webpage "https://en.wikipedia.org/wiki/Raymond_Leane" for the contents and seealso section
# importing libraries
from bs4 import BeautifulSoup
import urllib.request
import re
```

```python
#Connect to the webpage
url = "https://en.wikipedia.org/wiki/Raymond_Leane"
page  = urllib.request.urlopen(url) 
```

```python
#Parse the HTML  and a beautiful soup object is created
parser = BeautifulSoup(page, 'html.parser')
print(parser)
```

```python
#Now to webscrape "content " section of webpage
#use the beautiful soup obejct and findall the sections as "tocsection-"
#find_al gives an array  of list items 
regex =  re.compile('^tocsection-')
content_list = parser.find_all('li',attrs = {'class' :regex})
print(content_list)
    
```

```python
#To get the raw data alone
content = []
for li in content_list:
    content.append(li.getText().split('\n')[0])
    
print(content)    
```

```python
#Now to scrape data from SeeAlso section
#use find to get the div fromlist items and thento get the array using find_all
see_also_section = parser.find('div',attrs = {'class':'reflist columns references-column-width'})
see_also_soup =  see_also_section.find_all('li')
print(see_also_soup)

```

```python
#To get the raw data
see_also = []
for li in see_also_soup:
    a_tag = li.find('a',href=True,attrs={'class_':False})
    href = a_tag['href']
    text = a_tag.getText()
    see_also.append([href,text]) # append to array
print(see_also)
    

    
```

```python
#To save the data to txt or csv
with open('content.txt', 'w') as f:
    for i in content:
        f.write(i+"\n")
```

```python
with open('see_also.csv', 'w') as f:
    for i in see_also:
        f.write(",".join(i)+"\n")
```
